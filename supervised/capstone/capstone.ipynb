```python
df_2 = pd.read_csv('all_sites_scores.csv')
df_2.head()
```

```python
df_2['ROTTEN_DIFF'] = df_2['RottenTomatoes']-df_2['RottenTomatoes_User']
df_2.head(8)
```

```python
# Calculating the avarage of user and critics review diffrence
df_2['ROTTEN_DIFF'].apply(abs).mean()
```

```python
sns.histplot(df_2,x = 'ROTTEN_DIFF',kde = True,bins = 20)
```

```python
df_2.nlargest(10,'ROTTEN_DIFF')
```

```python
plt.figure(figsize = (10,10))
sns.scatterplot(df_2, x = 'Metacritic_User',y='IMDB')
```

```python
df_2[(df_2['Metacritic_User']) > 7 & (df_2['IMDB'] < 5 )]
```

```python
df_final = pd.merge(df,df_2 ,on= 'FILM',how='inner')
df_final.head(15)
```

```python
# df_final.shape
df_final.describe().transpose()['max']
```

```python
# NOrmalising data
df_final['Rt_norm'] = np.round(df_final['RottenTomatoes']/20,1)
df_final['Rtu_norm'] = np.round(df_final['RottenTomatoes_User']/20,1)
df_final['Meta_norm'] = np.round(df_final['Metacritic']/20,1)
df_final['MetaU_norm'] = np.round(df_final['Metacritic_User']/2,1)
df_final['IMDB_norm'] = np.round(df_final['IMDB']/2,1)
```

```python
df_final.head()
```

```python
df_final.columns
```

```python
norm_data = df_final[['STARS','RATING','Rt_norm','Rtu_norm','Meta_norm','MetaU_norm','IMDB_norm']]
norm_data.head()
```

```python
plt.figure(figsize = (10,4),dpi = 200)
sns.kdeplot(data = norm_data,clip = [0,5],shade = True,palette='Set1')
```
